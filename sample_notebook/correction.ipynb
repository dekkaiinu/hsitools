```python
import os
import sys
sys.path.append(os.pardir)
import numpy as np
from hsitools.convert import nh9_to_array


hsi_path = '../sample/sample_hsi.nh9'

hsi = nh9_to_array(hsi_path)
```

### hsi_blur()
HS画像に対して平滑化処理

```python
import numpy as np
from hsitools.correction import hsi_blur

smooth_hsi = hsi_blur(hsi, 50)
```

```python
import matplotlib.pyplot as plt
from hsitools.convert import hs_to_rgb, gamma_correction
hsi = hsi / 4095
rgb_img  = hs_to_rgb(hsi)
rgb_img  = gamma_correction(rgb_img, gamma=1.9, max_value=255)

plt.axis('off')
plt.imshow(rgb_img)
```

```python
smooth_rgb_img  = hs_to_rgb(smooth_hsi)
smooth_rgb_img  = gamma_correction(smooth_rgb_img, gamma=1.9, max_value=255)

plt.axis('off')
plt.imshow(smooth_rgb_img)
```
